***DATA VISULIZATION:***

***1.Best Art Style:***

```python
df['Art_Style'].value_counts()[:].plot(kind='line',figsize=(20,10),color='purple',title='Art Style').grid(color = 'brown',linewidth =0.7)
```

***2.Various Region:***

```python
df.groupby('Region')['Artwork_ID'].count().nlargest(8).plot(kind = 'pie')
```

***3.Checking Popularity Score with Density:***

```python
sns.distplot(df['Popularity_Score'])
font = {'family':'serif','color':'brown','size':20}
plt.title('Popularity Score',fontdict=font)
plt.show()
```

***4.Checking how much Tools are using in Art Style :***

```python
sns.violinplot(x= df['Tools_Used'], y= df['Art_Style'])
plt.show()
```

***5.Checking the Type of Genre in Art:***

```python
ig =plt.subplots(figsize =(11,10))
df['Art_Genre'].value_counts().head(10).plot(kind='pie', autopct = '%1.1f%%')
plt.legend(title = "Art Genre:")
```

***6.Checking the Score of Region:***

```python
art_type= df.groupby(['Region','Popularity_Score'])['Artwork_ID'].count()
```

```python
art_type
```

```python
art_type.plot(kind= 'bar', figsize= (30,15))
plt.title("Art Order",size=25)
```

***7.Type of Tools:***

```python
df['Tools_Used'].value_counts()[:20].plot(kind='bar', figsize=(20,7))
plt.title('Types of Tools used in Art by AI')
```

```python
df.head(2)
```

***8.No of Arts in a Region:***

```python
plt.figure(figsize=(15,5))
a=df['Art_Style'].value_counts()[:15]
sns.barplot(x=a,y=a.index,palette='rainbow')
plt.title('Number of Style of Arts')
plt.xlabel("Number of Arts",size=15)
plt.ylabel("Art_Style",size=15)
```

***9.Checking the count of various Platform:***

```python
df["Platform"].value_counts()
```

```python
plt.subplots(figsize=(10,8))
sns.countplot(x ='Platform', data = df, palette='spring')
plt.title('Various Platform')
```

***10.Checking the Score by ID:***

```python
plt.figure(figsize=(20,25))
sns.scatterplot(x='Popularity_Score',y='Artwork_ID',data=df)
plt.xlabel('Score',size=20)
plt.ylabel('ID',size=25)
```
